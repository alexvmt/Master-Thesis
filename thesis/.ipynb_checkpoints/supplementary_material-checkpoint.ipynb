#### Plot of visits, purchases, page views and product views per week

```python
# load data to generate plot
plot_data = pd.read_pickle('../data/processed_data/clickstream_0516-1016_prepared.pkl.gz')
```

```python
# create week column
plot_data['week'] = plot_data['visit_start_time_gmt'].apply(lambda x: x.isocalendar()[1])

# calculate visits per week
visits_per_week = plot_data.groupby('week').size().reset_index(name='visits')

# calculate purchases within next 24 hours per week
purchases_within_next_24_hours_per_week = plot_data[plot_data['purchase_within_next_24_hours'] == 1].groupby('week').size().reset_index(name='purchases_24_hours')

# calculate page views per week
page_views_per_week = plot_data.groupby('week', as_index=False).agg({'page_view_boolean_sum' : 'sum'})

# calculate product views per week
product_views_per_week = plot_data.groupby('week', as_index=False).agg({'product_view_boolean_sum' : 'sum'})
```

```python
# Figure 1
plt.figure(figsize=(12,8))

# visits per week
plt.subplot(2,2,1)
plt.plot(visits_per_week.week, visits_per_week.visits/1000, linewidth=2, marker='X', markersize=8)
plt.title('Visits', fontsize=16)
plt.ylabel('Number\n(thousands)', fontsize=16)
#plt.xlabel('Week', fontsize=16)
plt.tick_params(axis='y', labelsize=16)
plt.tick_params(axis='x', labelsize=16)
plt.grid(True)

# purchases per week
plt.subplot(2,2,2)
plt.plot(purchases_within_next_24_hours_per_week.week, purchases_within_next_24_hours_per_week.purchases_24_hours/1000, linewidth=2, marker='X', markersize=8)
plt.title('Purchases', fontsize=16)
#plt.ylabel('Number\n(thousands)', fontsize=16)
#plt.xlabel('Week', fontsize=16)
plt.tick_params(axis='y', labelsize=16)
plt.tick_params(axis='x', labelsize=16)
plt.grid(True)

# page views per week
plt.subplot(2,2,3)
plt.plot(page_views_per_week.week, page_views_per_week.page_view_boolean_sum/1000, linewidth=2, marker='X', markersize=8)
plt.title('Page views', fontsize=16)
plt.ylabel('Number\n(thousands)', fontsize=16)
plt.xlabel('Week', fontsize=16)
plt.tick_params(axis='y', labelsize=16)
plt.tick_params(axis='x', labelsize=16)
plt.grid(True)

# product views per week
plt.subplot(2,2,4)
plt.plot(product_views_per_week.week, product_views_per_week.product_view_boolean_sum/1000, linewidth=2, marker='X', markersize=8)
plt.title('Product views', fontsize=16)
#plt.ylabel('Number\n(thousands)', fontsize=16)
plt.xlabel('Week', fontsize=16)
plt.tick_params(axis='y', labelsize=16)
plt.tick_params(axis='x', labelsize=16)
plt.grid(True)

plt.tight_layout()
plt.savefig('visits_purchases_page_views_product_views_per_week.png')
plt.show()
```

#### AUC ranking

```python
# compute ranks based on AUC for all models and samples
learning_curves_auc_rank = learning_curves_auc.drop(['unique_visitors_train', 'unique_visitors (thousands)'], axis=1).rank(axis=1, ascending=False)

for model in ['LR', 'DT', 'NB', 'KNN', 'RF', 'SVM', 'BOOST', 'NN1', 'NN3', 'NN5', 'RNN', 'LSTM']:
    learning_curves_auc_rank[model] = learning_curves_auc_rank[model].apply(lambda x: int(x) if pd.notnull(x) else x)
    learning_curves_auc_rank.at[10, model] = learning_curves_auc_rank[model].mean()

learning_curves_auc_rank['unique_visitors (thousands)'] = learning_curves_auc['unique_visitors (thousands)']
learning_curves_auc_rank.at[10, 'unique_visitors (thousands)'] = 'mean rank'

learning_curves_auc_rank
```

```python
# get average rank based on AUC for each model
auc_rank = pd.DataFrame(index=['LR', 'DT', 'NB', 'KNN', 'RF', 'SVM', 'BOOST', 'NN1', 'NN3', 'NN5', 'RNN', 'LSTM'])
auc_rank['Mean rank'] = learning_curves_auc_rank.iloc[10]
auc_rank.sort_values('Mean rank', ascending=True, inplace=True)
auc_rank['Mean rank'] = auc_rank['Mean rank'].apply(lambda x: str(x)[:3])
auc_rank.at['KNN', 'Mean rank'] = str(11.3)
auc_rank
```

#### AUC ranking Friedman test

```python
# compute average rank based on AUC
print('Average rank based on AUC: ', auc_rank['Mean rank'].astype(np.float64).mean())

# compute chi squared value for F-test
F_Chi_squared = (12*10)/(12*(12+1))*(sum(auc_rank['Mean rank'].astype(np.float64)**2)-((12*((12+1)**2))/4))
print('Chi squared value for F-test: ', F_Chi_squared)

# compute F-value for F-test
F_F = ((10-1)*F_Chi_squared)/(10*(12-1)-F_Chi_squared)
print('F-value for F-test: ', F_F)

# compute critical distance for Nemenyi test
from math import sqrt
CD = 3.268003591*(sqrt((12*(12+1))/(6*10)))
print('Critical distance for Nemenyi test: ', CD)

# critical value for F-test: https://www.danielsoper.com/statcalc/calculator.aspx?id=4
# critical value for Nemenyi test: https://kourentzes.com/forecasting/2014/05/01/critical-values-for-the-nemenyi-test/
```

#### F-score ranking

```python
# compute ranks based on F-score for all models and samples
learning_curves_f_score_rank = learning_curves_f_score.drop(['unique_visitors_train', 'unique_visitors (thousands)'], axis=1).rank(axis=1, ascending=False)

for model in ['LR', 'DT', 'NB', 'KNN', 'RF', 'SVM', 'BOOST', 'NN1', 'NN3', 'NN5', 'RNN', 'LSTM']:
    learning_curves_f_score_rank[model] = learning_curves_f_score_rank[model].apply(lambda x: int(x) if pd.notnull(x) else x)
    learning_curves_f_score_rank.at[10, model] = learning_curves_f_score_rank[model].mean()

learning_curves_f_score_rank['unique_visitors (thousands)'] = learning_curves_f_score['unique_visitors (thousands)']
learning_curves_f_score_rank.at[10, 'unique_visitors (thousands)'] = 'mean rank'

learning_curves_f_score_rank
```

```python
# get average rank based on F-score for each model
f_score_rank = pd.DataFrame(index=['LR', 'DT', 'NB', 'KNN', 'RF', 'SVM', 'BOOST', 'NN1', 'NN3', 'NN5', 'RNN', 'LSTM'])
f_score_rank['Mean rank'] = learning_curves_f_score_rank.iloc[10]
f_score_rank.sort_values('Mean rank', ascending=True, inplace=True)
f_score_rank['Mean rank'] = f_score_rank['Mean rank'].apply(lambda x: str(x)[:3])
f_score_rank.at['NB', 'Mean rank'] = str(11.8)
f_score_rank.at['KNN', 'Mean rank'] = str(10.5)
f_score_rank
```

#### F-score ranking Friedman test

```python
# compute average rank based on F-score
print('Average rank based on F-score: ', f_score_rank['Mean rank'].astype(np.float64).mean())

# compute chi square value for F-test
F_Chi_squared = (12*10)/(12*(12+1))*(sum(f_score_rank['Mean rank'].astype(np.float64)**2)-((12*((12+1)**2))/4))
print('Chi square value for F-test: ', F_Chi_squared)

# compute F-value for F-test
F_F = ((10-1)*F_Chi_squared)/(10*(12-1)-F_Chi_squared)
print('F-value for F-test: ', F_F)

# compute critical distance for Nemenyi test
from math import sqrt
CD = 3.268003591*(sqrt((12*(12+1))/(6*10)))
print('Critical distance for Nemenyi test: ', CD)

# critical value for F-test: https://www.danielsoper.com/statcalc/calculator.aspx?id=4
# critical value for Nemenyi test: https://kourentzes.com/forecasting/2014/05/01/critical-values-for-the-nemenyi-test/
```

#### Ranks for all models and metrics

```python
# get ranks for all models and metrics and compute difference between a model's ranks
ranks = pd.merge(auc_rank, f_score_rank, suffixes=(' (AUC)', (' (F-score)')), how='inner', left_index=True, right_index=True).sort_index()
ranks['Mean rank (AUC)'] = ranks['Mean rank (AUC)'].astype(np.float64)
ranks['Mean rank (F-score)'] = ranks['Mean rank (F-score)'].astype(np.float64)
ranks['Absolute difference'] = abs(ranks['Mean rank (AUC)']-ranks['Mean rank (F-score)'])
pd.options.display.float_format = '{:20,.1f}'.format
ranks
```

```python
# gather training and testing times from model performance dataframes
times = model_performance_4k[['Training time', 'Testing time']].rename(columns={'Training time' : 'Sample 1 training time', 'Testing time' : 'Sample 1 testing time'})
samples = [2, 3, 4, 5, 6, 7, 8, 9, 10]
model_performances = [model_performance_8k, 
                      model_performance_16k, 
                      model_performance_31k, 
                      model_performance_63k, 
                      model_performance_125k, 
                      model_performance_250k, 
                      model_performance_500k, 
                      model_performance_1000k, 
                      model_performance_2000k]
for sample,model_performance in zip(samples,model_performances):
    times = pd.concat([times, model_performance[['Training time', 'Testing time']].rename(columns={'Training time' : 'Sample '+str(sample)+' training time', 'Testing time' : 'Sample '+str(sample)+' testing time'})], axis=1, sort=False)

# create dataframes for training and testing times
index = ['Training set 1', 
         'Training set 2', 
         'Training set 3', 
         'Training set 4', 
         'Training set 5', 
         'Training set 6', 
         'Training set 7', 
         'Training set 8', 
         'Training set 9', 
         'Training set 10']
train_times = pd.DataFrame(index=index)
train_times['unique_visitors'] = [3125, 6250, 12500, 25000, 50000, 100000, 200000, 400000, 800000, 1600000]

index = ['Test set 1', 
         'Test set 2', 
         'Test set 3', 
         'Test set 4', 
         'Test set 5', 
         'Test set 6', 
         'Test set 7', 
         'Test set 8', 
         'Test set 9', 
         'Test set 10']
test_times = pd.DataFrame(index=index)
test_times['unique_visitors'] = [781, 1502, 3125, 6250, 12500, 25000, 50000, 100000, 200000, 400000]

# load training and testing times for all samples and models into respective dataframes
models = ['LR', 'DT', 'NB', 'KNN', 'RF', 'SVM', 'BOOST', 'NN1', 'NN3', 'NN5', 'RNN', 'LSTM']
for model in models:
    train_times[model] = [model_performance_4k.at[model,'Training time'], 
                       model_performance_8k.at[model,'Training time'], 
                       model_performance_16k.at[model,'Training time'], 
                       model_performance_31k.at[model,'Training time'], 
                       model_performance_63k.at[model,'Training time'], 
                       model_performance_125k.at[model,'Training time'], 
                       model_performance_250k.at[model,'Training time'], 
                       model_performance_500k.at[model,'Training time'], 
                       model_performance_1000k.at[model,'Training time'], 
                       model_performance_2000k.at[model,'Training time']]    
for model in models:
    test_times[model] = [model_performance_4k.at[model,'Testing time'], 
                       model_performance_8k.at[model,'Testing time'], 
                       model_performance_16k.at[model,'Testing time'], 
                       model_performance_31k.at[model,'Testing time'], 
                       model_performance_63k.at[model,'Testing time'], 
                       model_performance_125k.at[model,'Testing time'], 
                       model_performance_250k.at[model,'Testing time'], 
                       model_performance_500k.at[model,'Testing time'], 
                       model_performance_1000k.at[model,'Testing time'], 
                       model_performance_2000k.at[model,'Testing time']]

# some modifications for plotting
train_times['unique_visitors (thousands)'] = train_times['unique_visitors'].apply(lambda x: str(int(x/1000)))
test_times['unique_visitors (thousands)'] = test_times['unique_visitors'].apply(lambda x: str(int(x/1000)))
train_times.at['Training set 3', 'unique_visitors (thousands)'] = '13'
test_times.at['Test set 1', 'unique_visitors (thousands)'] = '0.8'
test_times.at['Test set 2', 'unique_visitors (thousands)'] = '1.6'
test_times.at['Test set 5', 'unique_visitors (thousands)'] = '13'
```
