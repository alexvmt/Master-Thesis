#### Plot of visits, purchases, page views and product views per week

```python
# load data to generate plot
plot_data = pd.read_pickle('../data/processed_data/clickstream_0516-1016_prepared.pkl.gz')
```

```python
# create week column
plot_data['week'] = plot_data['visit_start_time_gmt'].apply(lambda x: x.isocalendar()[1])

# calculate visits per week
visits_per_week = plot_data.groupby('week').size().reset_index(name='visits')

# calculate purchases within next 24 hours per week
purchases_within_next_24_hours_per_week = plot_data[plot_data['purchase_within_next_24_hours'] == 1].groupby('week').size().reset_index(name='purchases_24_hours')

# calculate page views per week
page_views_per_week = plot_data.groupby('week', as_index=False).agg({'page_view_boolean_sum' : 'sum'})

# calculate product views per week
product_views_per_week = plot_data.groupby('week', as_index=False).agg({'product_view_boolean_sum' : 'sum'})
```

```python
# Figure 1
plt.figure(figsize=(12,8))

# visits per week
plt.subplot(2,2,1)
plt.plot(visits_per_week.week, visits_per_week.visits/1000, marker='x', markersize=8)
plt.title('Visits per week')
plt.ylabel('Number of visits\n(in thousands)')
plt.xlabel('Week')
plt.grid(True)

# purchases per week
plt.subplot(2,2,2)
plt.plot(purchases_within_next_24_hours_per_week.week, purchases_within_next_24_hours_per_week.purchases_24_hours/1000, marker='x', markersize=8)
plt.title('Purchases per week')
plt.ylabel('Number of purchases\n(in thousands)')
plt.xlabel('Week')
plt.grid(True)

# page views per week
plt.subplot(2,2,3)
plt.plot(page_views_per_week.week, page_views_per_week.page_view_boolean_sum/1000, marker='x', markersize=8)
plt.title('Page views per week')
plt.ylabel('Number of page views\n(in thousands)')
plt.xlabel('Week')
plt.grid(True)

# product views per week
plt.subplot(2,2,4)
plt.plot(product_views_per_week.week, product_views_per_week.product_view_boolean_sum/1000, marker='x', markersize=8)
plt.title('Product views per week')
plt.ylabel('Number of product views\n(in thousands)')
plt.xlabel('Week')
plt.grid(True)

plt.tight_layout()
plt.savefig('visits_purchases_page_views_product_views_per_week.png')
plt.show()
```

```python
# gather training and testing times from model performance dataframes
times = model_performance_4k[['Training time', 'Testing time']].rename(columns={'Training time' : 'Sample 1 training time', 'Testing time' : 'Sample 1 testing time'})
samples = [2, 3, 4, 5, 6, 7, 8, 9, 10]
model_performances = [model_performance_8k, 
                      model_performance_16k, 
                      model_performance_31k, 
                      model_performance_63k, 
                      model_performance_125k, 
                      model_performance_250k, 
                      model_performance_500k, 
                      model_performance_1000k, 
                      model_performance_2000k]
for sample,model_performance in zip(samples,model_performances):
    times = pd.concat([times, model_performance[['Training time', 'Testing time']].rename(columns={'Training time' : 'Sample '+str(sample)+' training time', 'Testing time' : 'Sample '+str(sample)+' testing time'})], axis=1, sort=False)

# create dataframes for training and testing times
index = ['Training set 1', 
         'Training set 2', 
         'Training set 3', 
         'Training set 4', 
         'Training set 5', 
         'Training set 6', 
         'Training set 7', 
         'Training set 8', 
         'Training set 9', 
         'Training set 10']
train_times = pd.DataFrame(index=index)
train_times['unique_visitors'] = [3125, 6250, 12500, 25000, 50000, 100000, 200000, 400000, 800000, 1600000]

index = ['Test set 1', 
         'Test set 2', 
         'Test set 3', 
         'Test set 4', 
         'Test set 5', 
         'Test set 6', 
         'Test set 7', 
         'Test set 8', 
         'Test set 9', 
         'Test set 10']
test_times = pd.DataFrame(index=index)
test_times['unique_visitors'] = [781, 1502, 3125, 6250, 12500, 25000, 50000, 100000, 200000, 400000]

# load training and testing times for all samples and models into respective dataframes
models = ['LR', 'DT', 'NB', 'KNN', 'RF', 'SVM', 'BOOST', 'NN1', 'NN3', 'NN5', 'RNN', 'LSTM']
for model in models:
    train_times[model] = [model_performance_4k.at[model,'Training time'], 
                       model_performance_8k.at[model,'Training time'], 
                       model_performance_16k.at[model,'Training time'], 
                       model_performance_31k.at[model,'Training time'], 
                       model_performance_63k.at[model,'Training time'], 
                       model_performance_125k.at[model,'Training time'], 
                       model_performance_250k.at[model,'Training time'], 
                       model_performance_500k.at[model,'Training time'], 
                       model_performance_1000k.at[model,'Training time'], 
                       model_performance_2000k.at[model,'Training time']]    
for model in models:
    test_times[model] = [model_performance_4k.at[model,'Testing time'], 
                       model_performance_8k.at[model,'Testing time'], 
                       model_performance_16k.at[model,'Testing time'], 
                       model_performance_31k.at[model,'Testing time'], 
                       model_performance_63k.at[model,'Testing time'], 
                       model_performance_125k.at[model,'Testing time'], 
                       model_performance_250k.at[model,'Testing time'], 
                       model_performance_500k.at[model,'Testing time'], 
                       model_performance_1000k.at[model,'Testing time'], 
                       model_performance_2000k.at[model,'Testing time']]

# some modifications for plotting
train_times['unique_visitors (in thousands)'] = train_times['unique_visitors'].apply(lambda x: str(int(x/1000)))
test_times['unique_visitors (in thousands)'] = test_times['unique_visitors'].apply(lambda x: str(int(x/1000)))
test_times.at['Test set 1', 'unique_visitors (in thousands)'] = '0.7'
```
